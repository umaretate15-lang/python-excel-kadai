```python
from datetime import datetime

class Event:
    # コンストラクタ
    def __init__(self, name, date, location):
        self.name = name
        self.date = date  # 文字列で受け取る（例: "2026-12-01"）
        self.location = location

    # 開催日までの日数を返すメソッド
    def days_until(self):
        today = datetime.now()
        event_date = datetime.strptime(self.date, "%Y-%m-%d")
        delta = event_date - today
        return delta.days


# ------------------------
# インスタンス作成
# ------------------------
event = Event(
    "プログラミング初心者向けセミナー",
    "2026-12-01",  # 未来の日付にする
    "東京国際フォーラム"
)

# ------------------------
# 属性の出力
# ------------------------
print(event.name)
print(event.date)
print(event.location)

# ------------------------
# 残り日数の出力
# ------------------------
print(f"開催日まであと{event.days_until()}日")
```

```python
class Human:
    # コンストラクタ
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # 大人かどうか判定するメソッド
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です")
        else:
            print(f"{self.name}は大人ではありません")


# ------------------------
# インスタンスをリストに追加
# ------------------------
humans = [
    Human("侍太郎", 25),
    Human("侍花子", 18),
    Human("侍次郎", 30)
]

# ------------------------
# 繰り返し処理でメソッド実行
# ------------------------
for human in humans:
    human.check_adult()
```

```python
from openpyxl import Workbook
from openpyxl.styles import Alignment, Font, Border, Side

# 新規ブック作成
wb = Workbook()
ws = wb.active

# 列幅調整
ws.column_dimensions['A'].width = 15
ws.column_dimensions['B'].width = 10
ws.column_dimensions['C'].width = 10
ws.column_dimensions['D'].width = 12
ws.column_dimensions['E'].width = 12

# タイトル
ws['B2'] = "請求書"
ws['B2'].font = Font(size=16, bold=True)

# 固定データ
ws['A4'] = "株式会社ABC"
ws['A5'] = "〒101-0022 東京都千代田区神田練塀町300"
ws['A6'] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws['A7'] = "担当者名:鈴木一郎 様"

# 右側情報
ws['E4'] = "No."
ws['F4'] = "0001"
ws['E5'] = "日付"
ws['F5'] = "2023/04/08"

# 表ヘッダー
ws['A10'] = "商品名"
ws['B10'] = "数量"
ws['C10'] = "単価"
ws['D10'] = "金額"

# データ
ws['A11'] = "商品A"
ws['B11'] = 2
ws['C11'] = 10000
ws['D11'] = 20000

ws['A12'] = "商品B"
ws['B12'] = 1
ws['C12'] = 15000
ws['D12'] = 15000

# 合計
ws['C13'] = "合計"
ws['D13'] = 35000

ws['C15'] = "消費税"
ws['D15'] = 3500

ws['C16'] = "税込合計"
ws['D16'] = 38500

# 枠線設定
thin = Border(
    left=Side(style='thin'),
    right=Side(style='thin'),
    top=Side(style='thin'),
    bottom=Side(style='thin')
)

for row in ws.iter_rows(min_row=10, max_row=12, min_col=1, max_col=4):
    for cell in row:
        cell.border = thin
        cell.alignment = Alignment(horizontal='center')

# 保存
wb.save("請求書.xlsx")
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Step1 & Step2: Excel読み込み
df_2022 = pd.read_excel('2022_年間売上表.xlsx')
df_2023 = pd.read_excel('2023_年間売上表.xlsx')

# Step3: データ連結
df_all = pd.concat([df_2022, df_2023], ignore_index=True)

# 商品 × 売上年ごとに集計
df_summary = df_all.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()

# Step4: Excel書き込み
output_file = '売上集計表.xlsx'
df_summary.to_excel(output_file, index=False)

# openpyxlでヘッダーの色変更
wb = load_workbook(output_file)
ws = wb.active

# 薄いグレーの塗りつぶし
fill = PatternFill(start_color='F2F2F2', end_color='F2F2F2', fill_type='solid')

# 1行目（ヘッダー）に適用
for cell in ws[1]:
    cell.fill = fill

# 保存
wb.save(output_file)

print("売上集計表.xlsx を作成しました！")
```
